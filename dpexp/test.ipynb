```python
#Read all images and then generate angles based on the center of mass of each circle.
import math
import time
import matplotlib.pyplot as plt
import cv2
import numpy as np
state = [[] for i in range(480)]
plt.rcParams["figure.figsize"] = (10,10)

fig,ax = plt.subplots(1,2)
ax1,ax2 = ax[0],ax[1]
FRAME_RATE = 1.0/60.0

for i in range(1,480):
    img = cv2.imread(f"images/aidenexp3/frame{i}.jpg")
    b,g,r = cv2.split(img)
    b = cv2.erode(b , np.ones((5, 5), np.uint8))
    g = cv2.erode(g , np.ones((5, 5), np.uint8))
    r = cv2.erode(r , np.ones((5, 5), np.uint8))

    ret,bmask = cv2.threshold(b,230,255,cv2.THRESH_BINARY)
    ret,gmask = cv2.threshold(g,230,255,cv2.THRESH_BINARY)
    ret,rmask = cv2.threshold(r,230,255,cv2.THRESH_BINARY)

    bcenter = cv2.moments( bmask, binaryImage=True)
    rcenter = cv2.moments( rmask, binaryImage=True)
    gcenter = cv2.moments( gmask, binaryImage=True)
    bcenter = (int(bcenter["m10"] / bcenter["m00"]),int(bcenter["m01"] / bcenter["m00"]) )
    rcenter = (int(rcenter["m10"] / rcenter["m00"]),int(rcenter["m01"] / rcenter["m00"]) )
    gcenter = (int(gcenter["m10"] / gcenter["m00"]),int(gcenter["m01"] / gcenter["m00"]) )

    l1 = np.sqrt((bcenter[0] - rcenter[0])**2 + (bcenter[1] - rcenter[1])**2)
    l2 = np.sqrt((rcenter[0] - gcenter[0])**2 + (rcenter[1] - gcenter[1])**2)
    # ax[0].imshow(img)
    # ax[1].imshow(bmask)
    # fig.show()
    # time.sleep(10)
    x1,y1 = bcenter[0]- rcenter[0], bcenter[1] - rcenter[1]
    x2,y2 = rcenter[0] - gcenter[0], rcenter[1] - gcenter[1]
    
    t1 = np.arctan2(-1,0) - np.arctan2(y1,x1) 
    t2 = np.arctan2(-1,0) - np.arctan2(y2,x2) 
    if(i ==1):
        state[i] = [t1, 0, t2, 0]
    else:
        state[i] = [t1, (t1- state[i-1][0])/FRAME_RATE ,t2,  (t2- state[i-1][2])/FRAME_RATE]
    if(sum([math.isnan(i) for i in state[i]]) >= 1):
        ax[0].imshow(img)
        fig.show()
        print(state[i])
        break
# for i in range(480):
state = state[1:]

```

```python
import scipy
from scipy.optimize import minimize
from scipy import integrate

G = 9.8  # acceleration due to gravity, in m/s^2
# L1 = 1.0  # length of pendulum 1 in m
# L2 = 1.0  # length of pendulum 2 in m
# M1 = 1.0  # mass of pendulum 1 in kg
# M2 = 1.0  # mass of pendulum 2 in kg
# change of energy over change of time.
#state = th1 th2 w10 w20 damp1 damp2\
def calcEnergy():
    pass
def derivs(t,state, L2,M2,a,b,L1= 1.0, M1= 1.0):
    dydx = np.zeros_like(state)
    dydx[0] = state[1]

    delta = state[2] - state[0]
    den1 = (M1+M2) * L1 - M2 * L1 * np.cos(delta) * np.cos(delta)
    dydx[1] = ((M2 * L1 * state[1] * state[1] * np.sin(delta) * np.cos(delta)
                + M2 * G * np.sin(state[2]) * np.cos(delta)
                + M2 * L2 * state[3] * state[3] * np.sin(delta)
                - (M1+M2) * G * np.sin(state[0]))
               / den1) + a*t
    

    dydx[2] = state[3]

    den2 = (L2/L1) * den1
    dydx[3] = ((- M2 * L2 * state[3] * state[3] * np.sin(delta) * np.cos(delta)
                + (M1+M2) * G * np.sin(state[0]) * np.cos(delta)
                - (M1+M2) * L1 * state[1] * state[1] * np.sin(delta)
                - (M1+M2) * G * np.sin(state[2]))
               / den2) + b*t
   

    return np.array(dydx)
    
def currentState(initial, t, g):
    return scipy.integrate.odeint(lambda x,t: derivs(t,x, *g),initial, t)

def rescale(arr):
    for i in range(len(arr)):
        if arr[i] > 2*np.pi:
            arr[i] -= arr[i] // 2*np.pi
        if arr[i] < 0:
            arr[i] += arr[i] // 2*np.pi

def distances(x,y):
    rx1 = np.sin([i[0] for i in x])
    rx2 = np.sin([i[2] for i in x]) + rx1
    ry1 = -np.cos([i[0] for i in x])
    ry2 = -np.cos([i[2] for i in x]) + ry1
    ax1 = np.sin([i[0] for i in y])
    ax2 = np.sin([i[2] for i in y]) + rx1
    ay1 = -np.cos([i[0] for i in y])
    ay2 = -np.cos([i[2] for i in y]) + ry1
    return sum([np.sqrt(i[0]**2-i[1]**2) for i in [(rx1,ax1),(rx2,ax2),(ry1,ay1), (ry2,ay2)]] )

def error(g):
    s = 0
    res = currentState(state[106], [j*FRAME_RATE for j in range(0,len(state)-106)], g)

    arr = np.array(state[106:])
    s+= distances(arr,res)
    
    return s
scipy.optimize.minimize(error, [1.,1.,0,0], method="Nelder-Mead", options={"disp":True, "maxiter" : 5000})
```

```python
class KalmanFilter:
    qt = []
    def __init__(self,qt, rt, p0, g, h, currentState, ht):
        #Model covariance matrix
        self.qt = qt
        #Sample Covariance matrix
        self.rt = rt
        #estimate covariance
        self.p0 = p0
        #model parameters to predict
        self.g = g
        #time array
        self.t = [0]
        #observations
        self.h = h
        #Model function
        self.currentState = currentState
        #hypothesis jacobian
        self.ht = ht
        self.x1 = h[0]
    def predictUpdate(self):
        k = self.p0 * (self.ht.transpose()* np.linalg.inv(self.rt))
        self.x1 = self.currentState(self.x1, [0, FRAME_RATE]) + k * (self.h[len(self.t)-1] - self.currentState(self.x1, g, [0, FRAME_RATE] ))
        # calculate jacobian. We only know the derivatives on the diagonal here.
        #scipy.optimize.approx_fprime(x1,derivs)
        ft = scipy.optimize.approx_fprime(self.x1,lambda x: currentState(self.h[0]))

        self.p0 = (ft*self.p0) + (self.p0 * ft.transpose()) -  ((k*self.ht)*self.p0) + self.qt
        self.t.append(len(self.t) *FRAME_RATE)
        return self.x1,self.p0
```
